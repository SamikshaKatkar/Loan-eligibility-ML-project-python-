# Exploratory Data Analysis & Visualization

### Loan Applicants based on Categorical Features

```python
def Personal_Data():
    plt.figure(figsize=(8, 6))

    # Gender wise distribution
    plt.subplot(1,4,1)
    sns.countplot(x='Gender', data=df)
     
    # Education wise distribution
    plt.subplot(1,4,2)
    sns.countplot(x='Education',data=df)
    plt.xticks(rotation=25)
    
     # Self-Employed wise distribution 
    plt.subplot(1,4,3)
    sns.countplot(x='Self_Employed',data=df)
    
    # Property-Area wise distribution 
    plt.subplot(1,4,4)
    sns.countplot(x='Property_Area',data=df)
    plt.xticks(rotation=25)
    
    plt.tight_layout()
    plt.show()
    

Personal_Data()
```

### Distribution of Numerical Features

```python
def Professional_Data():
    plt.figure(figsize=(8,6))

    # Applicant Income wise distribution
    plt.subplot(1,2,1)
    plt.hist(df['ApplicantIncome'], bins=30, alpha=0.7, color='b')
    plt.title('Applicant Income Distribution')
    plt.axvline(df['ApplicantIncome'].mean(),color='red',label='mean')
    plt.axvline(df['ApplicantIncome'].median(),color='yellow',label='median')
    
    # Co-Applicant Income wise distribution
    plt.subplot(1,2,2)
    plt.hist(df['CoapplicantIncome'], bins=30, alpha=0.7, color='b')
    plt.title('Co-Applicant Income Distribution')
    plt.axvline(df['CoapplicantIncome'].mean(),color='red',label='mean')
    plt.axvline(df['CoapplicantIncome'].median(),color='yellow',label='median')
    
    plt.tight_layout()
    plt.show()
    

Professional_Data()
```

```python
def Bank_Data():
    plt.figure(figsize=(8,8))

    # Loan Amount wise distribution
    plt.subplot(1,2,1)
    plt.hist(df['LoanAmount'], bins=30, alpha=0.7, color='b')
    plt.title('Loan Amount Distribution')
    plt.axvline(df['LoanAmount'].mean(),color='red',label='mean')
    plt.axvline(df['LoanAmount'].median(),color='yellow',label='median')
    
    # Credit History wise distribution
    plt.subplot(1,2,2)
    plt.hist(df['Credit_History'], bins=30, alpha=0.7, color='b')
    plt.title('Credit History Distribution')
    plt.axvline(df['Credit_History'].mean(),color='red',label='mean')
    plt.axvline(df['Credit_History'].median(),color='yellow',label='median')
    
    plt.tight_layout()
    plt.show()
    

Bank_Data()
```

# Machine Learning Algorithms

### Train-Test Split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

### 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report
ac=accuracy_score(ytest,ypred)
cr=classification_report(ytest,ypred)
```

```python
print(ac)
```

```python
print(cr)
```

### Hypertunning

```python
logreg=LogisticRegression(solver='liblinear')
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
print(classification_report(ytest,ypred))
```

```python
logreg=LogisticRegression(solver='saga')
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
print(classification_report(ytest,ypred))
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
logreg=LogisticRegression(solver='saga')
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
print(classification_report(ytest,ypred))
```

### 2) Decision Tree Classifier

```python
from sklearn.svm import SVC
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
logreg=LogisticRegression()
svm=SVC()
dt=DecisionTreeClassifier()
```

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    
    as_=accuracy_score(ytest,ypred)
    cs=classification_report(ytest,ypred)
    
    print(as_)
    print(cr)
    
    return model
```

```python
mymodel(logreg)
```

```python
mymodel(svm)
```

```python
mymodel(dt)
```

```python
train=dt.score(xtrain,ytrain)
test=dt.score(xtest,ytest)
print(f"training score is {train} and testing score is {test}")
```

```python
fig,ax=plt.subplots(figsize=(25,10))
chart=tree.plot_tree(dt,max_depth=5,feature_names=x.columns.tolist(),filled=True,fontsize=10)
plt.savefig('loan.jpg')
```

### Hypertunning 

#### Max_Depth

```python
for i in range(1,30):
    dt1=DecisionTreeClassifier(max_depth=i)
    dt1.fit(xtrain,ytrain)
    ypred=dt1.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f"max_depth = {i} and accuracy = {ac}")
```

```python
dt2 = DecisionTreeClassifier(max_depth=3)
mymodel(dt2)
```

```python
train=dt2.score(xtrain,ytrain)
test=dt2.score(xtest,ytest)
print(f"Training Score {train} and Testing Score {test}")
```

#### Min_Samples_Leaf

```python
for i in range(31,60):
    dt3=DecisionTreeClassifier(min_samples_leaf=i)
    dt3.fit(xtrain,ytrain)
    ypred=dt3.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f"max_depth = {i} and accuracy = {ac}")
```

```python
dt4 = DecisionTreeClassifier(min_samples_leaf=55)
mymodel(dt4)
```

```python
train=dt4.score(xtrain,ytrain)
test=dt4.score(xtest,ytest)
print(f"Training Score {train} and Testing Score {test}")
```

#### Min_Samples_Split

```python
for i in range(100,130):
    dt5=DecisionTreeClassifier(min_samples_split=i)
    dt5.fit(xtrain,ytrain)
    ypred=dt5.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f"max_depth = {i} and accuracy = {ac}")
```

```python
dt6 = DecisionTreeClassifier(min_samples_split=321)
mymodel(dt6)
```

```python
train=dt6.score(xtrain,ytrain)
test=dt6.score(xtest,ytest)
print(f"Training Score {train} and Testing Score {test}")
```

# 3) Ensemble Learning

### Bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bg=BaggingClassifier(LogisticRegression())
bg.fit(xtrain,ytrain)
ypred=bg.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
bg=BaggingClassifier(DecisionTreeClassifier())
bg.fit(xtrain,ytrain)
ypred=bg.predict(xtest)
print(classification_report(ytest,ypred))
```
